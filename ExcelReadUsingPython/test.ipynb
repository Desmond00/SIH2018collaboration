```python
import openpyxl
from openpyxl import Workbook

book = openpyxl.load_workbook('data.xlsx')
bookTemp = Workbook()

sheet = book.active
sheetTemp = bookTemp.active

rowIndex = 5
while rowIndex <= 280:
    #Checking onlly first column
    cellData1 = sheet.cell(row=rowIndex, column=1).value;
    nextCellData1 = sheet.cell(row=rowIndex+1, column=1).value
    print(cellData1, end=" , ")
    sheetTemp.cell(row=rowIndex, column=1).value = cellData1
    
    
    for columnIndex in range(2, 8):
        cellData = sheet.cell(row=rowIndex, column=columnIndex).value
        nextCellData = sheet.cell(row=rowIndex+1, column=columnIndex).value 
        
        #For null/blank in next row's serial number concatinating the contents of the next row with the current one
        if(nextCellData1 == None and nextCellData != None):
            cellData += " "+nextCellData
            
        
        sheetTemp.cell(row=rowIndex, column=columnIndex).value = cellData
        print(cellData, end = " , ")
    print(sheet.cell(row=rowIndex, column=8).value)
    sheetTemp.cell(row=rowIndex, column=8).value = sheet.cell(row=rowIndex, column=8).value
    if(nextCellData1 == None):
        rowIndex += 1
        nextCellData1 = 2
    rowIndex += 1
    
bookTemp.save("temp.xlsx")
```
